```python
import pandas as pd

# veriyi GitHub üzerindeki güvenilir bir yedekten (CSV olarak) çekiyoruz.
url = "https://raw.githubusercontent.com/databricks/Spark-The-Definitive-Guide/master/data/retail-data/all/online-retail-dataset.csv"
print("Veri GitHub yedeğinden indiriliyor...")

# read_csv ile csv dosyasını okuyoruz ve değişkene atıyoruz.
df_raw = pd.read_csv(url)

print(f" {len(df_raw)} satır ham veri hafızaya yüklendi.")
display(df_raw.head())
```

```python
# Askıda kalan bozuk motoru(engine) kapatıp temizliyoruz.
engine.dispose()

# Tertemiz, yeni bir motor oluşturuyoruz (connection_uri hafızada zaten var)
engine = create_engine(connection_uri)

print(" Veritabanı bağlantısı sıfırlandı.")

# yükleme işlemini tekrar deniyoruz
table_name = 'raw_sales'

df_raw.to_sql(
    name=table_name,
    con=engine,
    if_exists='replace',
    index=False,
    chunksize=10000
)

print(f" Ham veri '{table_name}' tablosu yüklendi.")
```

```python
# df_clean adında df_raw'ın bir kopyasını oluşturduk.
df_clean = df_raw.copy()
```

```python
#isna() metodu ile boolean maskeleme yapıyoruz.
#sum() ile ise maskeleme sonucunda dönenenlerin toplamı, sonuç olarak kaç tane boş hücre olduğunu verir.
df_clean['CustomerID'].isna().sum()
```

```python
# .fillna() metodunu kullanarak müsteri ID'si boş müşterilerin CustomerID kısmını -1 ile dolduruyoruz.
df_clean['CustomerID'] = df_clean['CustomerID'].fillna(-1)

# Kontrol kısmı: Eksik(NaN) değer kaldı mı?
print("Kalan eksik CustomerID sayısı:", df_clean['CustomerID'].isna().sum())
```

```python
# customerID'leri -1 olan sonuçları filtreleme
misafir_musteriler = df_clean[df_clean['CustomerID'] == -1]

print(f"Toplam Misafir Müşteri İşlemi: {len(misafir_musteriler)}")
display(misafir_musteriler.head())
```

```python
# iptal edilen veya iade edilen siparişleri temizliyoruz.
df_clean = df_clean[df_clean['Quantity'] > 0].copy()
print(f"Kalan satır sayısı: {len(df_clean)}")
```

```python
# InvoiceNo dtype --> object
# InvoiceNo dtype --> datetime olmalı
df_clean.info()
```

```python
# Metin halindeki tarihleri gerçek dateTime formatına çeviriyoruz.
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'])

print(df_clean['InvoiceDate'].head())
```

```python
# Her bir satır için toplam tutarı hesaplıyoruz.
df_clean['TotalAmount'] = df_clean['Quantity'] * df_clean['UnitPrice']

display(df_clean[['Quantity', 'UnitPrice', 'TotalAmount']].head())
```

```python
# Müşteriler Tablosu (dim_customer)
# Sadece CustomerID sütununu seçtik, drop_duplicates() ile tekrar eden müşteri idleri sildik. copy() ile bağımsız bir tablo oluşturduk.
dim_customer = df_clean[['CustomerID']].drop_duplicates().copy()

print(f"Müşteri sayısı: {len(dim_customer)}")
display(dim_customer.head())
```

```python
# Ürünler Tablosu (dim_product)
dim_product = df_clean[['StockCode', 'Description', 'UnitPrice']].drop_duplicates().copy()

print(f"Ürün çeşidi sayısı: {len(dim_product)}")
display(dim_product.head())
```

```python
# Zaman Tablosu (dim_date)
# Önce benzersiz tarihleri ana tablodan çekiyoruz.
dim_date = df_clean[['InvoiceDate']].drop_duplicates().copy()

# O tarihlerin içinden özellikleri çıkarıyoruz.
dim_date['Year'] = dim_date['InvoiceDate'].dt.year
dim_date['Month'] = dim_date['InvoiceDate'].dt.month
dim_date['Day'] = dim_date['InvoiceDate'].dt.day

# dayofweek: Pazartesi(0) ... Cumartesi(5), Pazar(6) kabul eder.
dim_date['DayOfWeek'] = dim_date['InvoiceDate'].dt.dayofweek
#Bu gün hafta sonu mu? (Cumartesi=5, Pazar=6)
dim_date['IsWeekend'] = dim_date['DayOfWeek'].isin([5, 6])

print(f" Tarih sayısı: {len(dim_date)}")
display(dim_date.head())
```

```python
# Fact Tablosu (Fact Table)
fact_sales = df_clean[['InvoiceNo', 'StockCode', 'CustomerID', 'InvoiceDate', 'Quantity', 'UnitPrice', 'TotalAmount']].copy()

print(f"fact_sales tablosu satır sayısı: {len(fact_sales)}")
display(fact_sales.head())
```

```python
engine.dispose()
engine = create_engine(connection_uri)

# dim_customer(Müşteriler) tablosunu Neon (PostgreSQL) yükleme kısmı
dim_customer.to_sql(name='dim_customer', con=engine, if_exists='replace', index=False)
print("dim_customer tablosu yüklendi.")


# dim_product(Ürünler) tablosunu Neon (PostgreSQL) yükleme kısmı
dim_product.to_sql(name='dim_product', con=engine, if_exists='replace', index=False)
print("dim_product tablosu yüklendi.")


# dim_date(Tarih) tablosunu Neon (PostgreSQL) yükleme kısmı
dim_date.to_sql(name='dim_date', con=engine, if_exists='replace', index=False)
print("dim_date tablosu yüklendi.")


# Satışlar Gerçek (Fact) tablosu Neon (PostgreSQL) yükleme kısmı
# Bu tablo büyük olduğu için yine chunksize (paketleme) kullanıyoruz.
fact_sales.to_sql(name='fact_sales', con=engine, if_exists='replace', index=False, chunksize=10000)
print("fact_sales tablosu yüklendi.")
```
